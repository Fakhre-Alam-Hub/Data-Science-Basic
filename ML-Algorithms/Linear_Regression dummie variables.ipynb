```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv(r"C:\dataset\carprice.csv")
train
```

```python
%matplotlib inline
plt.figure(figsize=(8,8))
plt.grid(True)
plt.xlabel('Mileage', fontsize=15)
plt.ylabel('Sell Price($)', fontsize=15)
plt.scatter(train['Mileage'],train['Sell Price($)'], marker='x', color='red')
```

```python
df_Car_Model = pd.get_dummies(train['Car Model'])
df_Car_Model
```

```python
final_train = pd.concat([train,df_Car_Model], axis='columns')
final_train
```

```python
final = final_train.drop(['Car Model','Mercedez Benz C class'], axis = 'columns')
final
```

```python
X = final.drop('Sell Price($)', axis = 'columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
model = LinearRegression()
model
```

```python
model.fit(X,y)
```

```python
# price of mecedez benz that is 4 year old with mileage 45000

model.predict([[45000,4,0,0]])
```

```python
# price of bmw x5 that is 7 year old with mileage 86000

model.predict([[86000,7,0,1]])
```

```python
# accuracy of the model

f"{model.score(X,y)*100} %"
```
