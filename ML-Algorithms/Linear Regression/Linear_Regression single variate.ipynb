```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv(r"C:\dataset\train_Company_Earning.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
%matplotlib inline
fig= plt.figure(figsize=(10,8))
plt.scatter(data['Year'], data['Earning'], color = 'r', marker = '+')
plt.axis(xmin=1999,xmax= 2016, ymin=100, ymax=800)
plt.title('Company Earning')
plt.xlabel('<---Year--->')
plt.ylabel('Earning')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg
```

```python
reg.fit(data[['Year']], data['Earning'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
fig= plt.figure(figsize=(10,8))
plt.scatter(data['Year'], data['Earning'], color = 'r', marker = '+')
plt.axis(xmin=1999,xmax= 2016, ymin=100, ymax=800)
plt.title('Company Earning')
plt.xlabel('<---Year--->')
plt.ylabel('Earning')
plt.plot(data['Year'], reg.predict(data[['Year']]), color='b')
plt.annotate("Y=mx + c (Best fit line)", xy=(2008, 350), xytext=(2006, 480),arrowprops=dict(arrowstyle="->"), fontsize=15)
```

```python
test = pd.read_csv(r'C:\dataset\test_Company_Earning.csv')
```
